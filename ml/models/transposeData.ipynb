```python
### 재료명 뽑아오기 -> 열 이름으로 활용
```

```python
import re

# 재료명 가져오기
igd_name = rcp_igd['재료명'].unique()

# 재료명에사 괄호 제거
for i in range(len(igd_name)):
    tmp = re.sub(r'[\[\]]', '', igd_name[i])
    tmp = tmp.replace(' ', '')
    igd_name[i] = tmp

# 정렬
igd_name = np.sort(igd_name)
igd_name = pd.Series(igd_name).unique()
print(igd_name)
```

```python
# 재료명 정리하기
# 규칙
# 
```

```python
igd_list = []
for i in range(0,len(rcp_igd['재료명'])):
    if rcp_igd['재료명'][i] in ["돛나물"]:
        rcp_id = rcp_igd.iloc[i,0];
        label = rcp_igd.iloc[i,0]
        for j in range(0,len(rcp_label)):
            if rcp_label.iloc[j,0] == label:
                idx = len(igd_list)
                igd_list.append(find_igd(rcp_id))
                igd_list[idx].insert(0,rcp_label.iloc[j,1])
print(igd_list)
```

```python
def find_igd(rcp_id):
    igd_list = []
    for i in range(0,len(rcp_igd['재료명'])):
        if rcp_igd.iloc[i,0] == rcp_id:
            igd_list += [rcp_igd.iloc[i,2]]
    return igd_list
```

```python
# DataFrame에서 특정 열의 고유한 값 가져오기
unique_recipe_names = rcp_label['레시피 이름'].unique()

# 고유한 레시피 이름을 텍스트 파일에 저장하기
file_path = "unique_recipe_names.txt"
with open(file_path, 'w') as file:
    for recipe_name in unique_recipe_names:
        file.write(recipe_name + '\n')

print("고유한 레시피 이름이 파일에 저장되었습니다:", file_path)
```
